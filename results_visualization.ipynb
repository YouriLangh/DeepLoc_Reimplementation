```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
df = pd.read_csv("results/DeepLoc1.0/train_val_metrics.csv") 

# Plot the two accuracy metrics
plt.figure(figsize=(10, 6))
plt.plot(df['epoch'], df['val_loc_acc'], label='Localization Accuracy (%)')
plt.plot(df['epoch'], df['val_mem_acc'], label='Membrane Accuracy (%)')

# Styling
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Validation Accuracy over Epochs")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
