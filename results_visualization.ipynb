```python
import pandas as pd
import matplotlib.pyplot as plt

# version = "DeepLoc1.0"
version = "DeepLoc2.0"
# Load the CSV data into a DataFrame
df = pd.read_csv("results/" + version + "/train_metrics.csv") 

# Plot the two accuracy metrics
plt.figure(figsize=(10, 6))
plt.plot(df['epoch'], df['loc_acc'], label='Localization Accuracy (%)')
plt.plot(df['epoch'], df['mem_acc'], label='Membrane Accuracy (%)')

# Styling
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy over Epochs")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the learning curve
plt.savefig("results/" + version + "/learning_curve.png", dpi=300)
# Show the plot
plt.show()
```

```python
# Plot localization and membrane loss over epochs
plt.figure(figsize=(10, 6))
plt.plot(df['epoch'], df['loc_loss'], label='Localization Loss')
plt.plot(df['epoch'], df['mem_loss'], label='Membrane Loss')

# Styling
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss over Epochs")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the loss curve
plt.savefig(f"results/{version}/loss_curve.png", dpi=300)
plt.show()
```
